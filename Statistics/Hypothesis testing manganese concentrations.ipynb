# Hypothesis test for  $\mu$

Imagine we're biologyst studying certain toxic substance level in a lake, our goal is to determine if these levels are low enough for it to be safe to fish there, of course we can take a sample and analyze it at the laboratory, what assures us that the sample is representative? how can we be sure that we aren't sampling a low-concentration zone, or it maybe the oposite and we're sampling in a high-concentration place. How to know it? How to take a decision? The best we can do is to sample many times throughout the lake and calculate the average.

But... Is this a definitive result? No, it isn't, sampling more certainly get us close to the "real concentration" but this is not 100% reliable, remember this is a "sample", if we wanted to get "real concentration" 100% realiable (that in fact is not a number perse but an scalar field) we should analyze the whole lake, which in practice is actually impossible because is expenssive and time consuming. 

What should we do then to get a reliable result that helps us in decision-making? In this case is convinient to perform a statistical test which ensures us with certain **confidence level** that our obtained result is close to the real one, this is called **Hypothesis test**.

### How to perform a Hypothesis Test?
Suppose we want to know if manganese concentration is lower than 2.5ppm, we've got to options:
- $H_0:$  $[Mn] \leq $ 2.5 ppm 
- $H_1:$ $[Mn] >$ 2.5 ppm  

This is how a hypothesis test is formulated, $H_0$ is called "null hypothesis" and is the one we accept by default, while $H_1$ is the alternative hypothesis and is that we try to prove.

### P-value
This is a statistic that in simple words represents the probability of the null hypothesis to be true, and this helps us to determine the result of the test, How? We generally fail against the null hypothesis if:

$$
p-value\geq 0.05
$$
And against the alternative one on the contrary.

### Nota: 
In order to apply this test must be fulfilled that our sample is **Normally distributed**

```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sb
from statsmodels.stats.weightstats import ztest as ztest
```

```python
# Creating the data
np.random.seed(2015)
concentraciones = np.random.normal(2.5,0.4,50)


# Ploting it
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.plot(concentraciones, linestyle='-', marker='o')
plt.title('Concentration vs Sample', size=20)
plt.xlabel('# of observation', size=13)
plt.ylabel('[ ppm ]', size=13)

plt.grid()

plt.subplot(1,2,2)
sb.kdeplot(x=concentraciones, fill=True, alpha=.8, linewidth=3)
plt.title('Concentration distribution', size=20)
plt.xlabel('[ ppm ]', size=13)
plt.grid()

plt.show()
```

```python
# Now we perform the z-test from the statsmodels library
Z, pval = ztest(concentraciones, value=2.5, alternative='smaller')
print('P-value =',round(pval,3))
```

```python
if pval < 0.05:
  print('The p-value is lower than 0.05, so we reject the null hypothesis')
else:
  print('The p-value is higher than 0.05, so we accept the null hypothesis')
```

The test concludes than the  manganase concentration is lower than 2.5ppm, so, the lake is safe to fish.
